```python
import numpy as np
import cv2
from matplotlib import pyplot as plt
```

```python
img = cv2.imread('dog.webp')
img
```

```python
plt.imshow(img)
```

```python
normal = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(normal)
```

```python
def getImage(img, cmap=None):
    plt.axis('off')
    plt.imshow(img, cmap=cmap)
    plt.show()
    
```

```python
getImage(normal)
```

```python
#colorized
hsv = cv2.cvtColor(normal, cv2.COLOR_BGR2HSV)
getImage(hsv)
```

```python
hsv = cv2.cvtColor(img, cv2.COLOR_BGR2HSV)
getImage(hsv)
```

```python
#gray scaled
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
getImage(gray, cmap='gray')
```

```python
width, heigth = normal.shape[:2]

if width > 500:
    k = (50,50)
elif width > 200 and width <= 500:
    k = (25,25)
else:
    k = (10,10)
```

```python
#blurred
blur = cv2.blur(normal, k)
getImage(blur)
```

```python
#Binary

v , binary = cv2.threshold(gray, 125, 255, cv2.THRESH_BINARY)
getImage(binary, cmap='gray')
```

```python
#invert 

invert = cv2.bitwise_not(binary)
getImage(invert, cmap='gray')
```
